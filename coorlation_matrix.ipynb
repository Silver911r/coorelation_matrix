```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn
import yfinance as yf
import plotly.express as px
```

```python
pd.set_option('display.max_columns', None, 'display.max_rows', None)
```

```python
#wikipedia link sp500 list
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
tables = pd.read_html(url)
```

```python
#pull sp500 list table out and save to sp500 data frame
sp500 = tables[0]
```

```python
#convert the data frame table of symbols into a list
sp500_symbols = sp500["Symbol"].to_list()
```

```python
#select start date for correlation window
start = datetime(2020, 1, 1)

#put in individual symbols, or pull sp500 and add any additional symbols desired
# symbols_list = ['F', 'TWTR', 'FB', 'AAL', 'AMZN', 'GOOGL', 'GE']
symbols_list = sp500_symbols
```

```python
#grab the data
data = yf.download(symbols_list, start)
```

```python
#save just the close prices to new data frame
data_close = data['Adj Close']
```

```python
#calculate the coorelation matrix
corr_df = data_close.corr(method='pearson')

corr_df.head(5)
```

```python
# enter stock symbol here to see a list of stock coorlations for that symbol
filtered_list = corr_df['AAPL']
```

```python
filtered_list[filtered_list > .96]
```

```python
# save an interactive html heatmap to look at all the data in one spot

fig = px.imshow(corr_df)
fig.write_html("heatmap.html")
```
